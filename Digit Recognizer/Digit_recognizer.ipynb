```python
import os
import cv2
import pandas as pd
import numpy as np
from zipfile import ZipFile

# Loading the XLS File
xls_file = "digit_recognizer.xlsx"
data = pd.read_excel(xls_file)

# Extract the Label and Pixel Values
labels = data.iloc[:, 0].values
pixels = data.iloc[:, 1:].values

# Reshape the Pixel Values
pixels = pixels.reshape(-1, 28, 28)

# Create Images and save Images
result_dir = "result"
os.makedirs(result_dir, exist_ok=True)

for i in range(len(labels)):
    label = str(labels[i])
    image = pixels[i].astype(np.uint8)
    image_path = os.path.join(result_dir, f"{label}.png")
    cv2.imwrite(image_path, image)

# Create a Zip file
zip_file = "result.zip"
with ZipFile(zip_file, "w") as zf:
    for image_file in os.listdir(result_dir):
        image_path = os.path.join(result_dir, image_file)
        zf.write(image_path, image_file)

print("Image conversion and saving completed.")
print(f"Result images saved in '{result_dir}' directory.")
print(f"Zip file '{zip_file}' created.")
```
